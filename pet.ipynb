```python
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from selenium import webdriver 
from selenium.webdriver.chrome.service import Service as ChromeService 



options = webdriver.ChromeOptions() #newly added 
options.add_argument('--headless')
driver = webdriver.Chrome(options=options)

state_name = []
city_name = []
clinic_name = []

# Open the file containing the JSON data
with open('json_file.txt', 'r') as file:
    # Read the contents of the file
    json_data = file.read()
    
    # Convert the JSON data from string to Python object
    parsed_data = json.loads(json_data)
    
    # Print the Python object
    #print(parsed_data)

    for item in parsed_data:
        state = item['state']
        
        for location in item['locationByStateAndCity']:
            city = location['city']

            for location_details in location['locations']:
                vanity_name = location_details['vanityName']
                #print(f'State: {state}, City: {city}, Vanity Name: {vanity_name}')
                state_name.append(state)
                city_name.append(city)
                clinic_name.append(vanity_name)
```

```python
c1 = [s.replace(' ', '-') for s in clinic_name]
c2 = [s.replace('---', '-') for s in c1]
c3 = [s.replace('&','and') for s in c2]
c4 = []

for i in range(0,387):
    url = f"https://www.thrivepetcare.com/locations/{state_name[i]}/{city_name[i]}/{c3[i]}"
    c4.append(url)


clinic_url = []
      
c5 = [s.replace(' ', '-') for s in c4]
c6 = [s.replace('---', '-') for s in c5]
c7 = [s.replace("'","-")for s in c6]
for c in c7:
      clinic_url.append(c.lower())

file = open('clinic_url.txt','w')
for link in clinic_url:
	file.write(link+"\n")
file.close()

```

```python

counter = 1

for l in clinic_url:
    driver.get(l)
    wait = WebDriverWait(driver, 10)
    title_result = driver.title
    if title_result == "404 | Thrive Pet Healthcare":
        print('false' , counter)
    counter = counter + 1
    
    

    
```

```python
Clinic_final_name = []
Clinic_offering = []
Clinic_timings = []
Clinic_add = []
Clinic_city = []
Clinic_state = []
Clinic_phone = []
Clinic_zip = []
Clinic_web = []
Clinic_hours = []
```

```python
import re

options = webdriver.ChromeOptions() #newly added
driver = webdriver.Chrome(options=options)

pattern = re.compile(r'\D+')

i = 0
for page in correct_urls:
    driver.get(page)
    time.sleep(5)

    offering = []

    try:
        name_w_loc = driver.find_element(By.XPATH, '//*[@id="hours"]/div/div/div[1]/h1')

        try:
            off_1 = driver.find_element(By.XPATH, '//*[@id="simple-tab-0"]')
            try:
                off_2 = driver.find_element(By.XPATH, '//*[@id="simple-tab-1"]')
                if off_2.text == "Primary Care" or off_2.text == "Specialty" or off_2.text == "Specialty Care" or off_2.text == "Emergency Care" or off_2.text == "Urgent Care":
                    try:
                        off_3 = driver.find_element(By.XPATH, '//*[@id="simple-tab-2"]')
                        if off_3.text == "Primary Care" or off_3.text == "Specialty" or off_3.text == "Specialty Care" or off_3.text == "Emergency Care" or off_3.text == "Urgent Care":
                            offering.append(off_1.text + '|' + off_2.text + '|' + off_3.text)
                        else:
                            offering.append(off_1.text + '|' + off_2.text)
                    except:
                        offering.append(off_1.text + '|'  + off_2.text)
                else:
                    offering.append(off_1.text)
            except:
                offering.append(off_1.text)
        except:
            off_1 = 'None'
            offering.append(off_1)

        try:   
            timings = driver.find_element(By.XPATH, '//*[@id="clinic-tabpanel-0"]/div/p/div/div[2]/p[2]/p')
            try:
                timings2 = driver.find_element(By.XPATH, '//*[@id="clinic-tabpanel-0"]/div/p/div/div[2]/p[3]/p')
                try:
                    timings3 = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/section/div/div/div[1]/div/div/div[2]/div/div/p/div/div[2]/p[4]/p')
                    clin_time = timings.text + "|" + timings2.text + timings3.text
                except:
                    clin_time = timings.text + "|" + timings2.text
            except:
                clin_time = timings.text
        except:
            clin_time = 'None'
        
        try:
            hours24 = driver.find_element(By.XPATH, '//*[@id="clinic-tabpanel-0"]/div/p/div/div[2]/p[2]')
            if hours24.text == "Open 24 Hours":
                hours_open = 'Yes'
            else:
                hours_open = 'No'
        except:
            hours_open = 'No'

        

        try:
            element = driver.find_element(By.XPATH, '//*[@id="hours"]/div/div/div[2]/div[2]/p')
            text1 = element.text.split(',')[0]
            text2 = element.text.split(',')[1]
            text3 = element.text.split(',')[3]
            zip = re.sub(pattern, '', text3)
            add = text1 + ',' + text2
        except:
            add = 'None'
            zip = 'None'

        try:
            phone = driver.find_element(By.XPATH, '//*[@id="Sect0_CTA10"]/p')
            pn = phone.text
        except:
            pn = 'None'
            
    except:
        name_w_loc = driver.find_element(By.XPATH, '//*[@id="__next"]/div[2]/div[1]/section[1]/div/div[1]/div/div[1]/div[1]/h1')

        try:
            off_1 = driver.find_element(By.XPATH, '//*[@id="__next"]/div[2]/div[1]/section[2]/div/div[1]/div[2]/div/div[2]/p')
            try:
                off_2 = driver.find_element(By.XPATH, '//*[@id="__next"]/div[2]/div[1]/section[2]/div/div[1]/div[2]/div[2]/div[2]/p')
                if off_2.text == "Primary Care" or off_2.text == "Specialty" or off_2.text == "Specialty Care" or off_2.text == "Emergency Care" or off_2.text == "Urgent Care":
                    try:
                        off_3 = driver.find_element(By.XPATH, '//*[@id="__next"]/div[2]/div[1]/section[2]/div/div[1]/div[2]/div[3]/div[2]/p')
                        if off_3.text == "Primary Care" or off_3.text == "Specialty" or off_3.text == "Specialty Care" or off_3.text == "Emergency Care" or off_3.text == "Urgent Care":
                            offering.append(off_1.text + '|' + off_2.text + '|' + off_3.text)
                        else:
                            offering.append(off_1.text + '|' + off_2.text)
                    except:
                        offering.append(off_1.text + '|'  + off_2.text)
                else:
                    offering.append(off_1.text)   
            except:
                offering.append(off_1.text)
        except:
            off_1 = 'None'

        try:
            timings = driver.find_element(By.XPATH, '//*[@id="__next"]/div[2]/div[1]/section[2]/div/div[2]/div[2]')
            clin_time = timings.text
        except:
            clin_time = 'None'

        try:
            element = driver.find_element(By.XPATH, '//*[@id="Sect1_CTA1"]/span')
            text1 = element.text.split(',')[0]
            text2 = element.text.split(',')[1]

            zip = re.sub(pattern, '', text2)
            add = text1
        except:
            zip = 'None'
            add = 'None'

        try:
            phone = driver.find_element(By.XPATH, '//*[@id="Sect1_CTA2"]')
            pn = phone.text
        except:
            pn = 'None'



    final_state = state_name[i]
    final_city = city_name[i]
    weblink = page

    print(name_w_loc.text)
    print(offering)
    print(clin_time)
    print(add) 
    print(zip)
    print(pn)
    print(final_state, final_city)
    print(weblink)
    print(hours_open)

    Clinic_final_name.append(name_w_loc.text)
    Clinic_offering.append(offering)
    Clinic_timings.append(clin_time)
    Clinic_add.append(add)
    Clinic_city.append(final_city)
    Clinic_state.append(final_state)
    Clinic_phone.append(pn)
    Clinic_zip.append(zip)
    Clinic_web.append(weblink)
    Clinic_hours.append(hours_open)

    i = i + 1
        
```

```python
data = pd.DataFrame({"Name of Clinic": Clinic_final_name,
 "Type of Offering":Clinic_offering,
  "Address":Clinic_add,
   "City":Clinic_city,
    "State":Clinic_state,
     "Zip":Clinic_zip,
      "Phone Number":Clinic_phone,
       "Hours":Clinic_timings,
        "Website Link":Clinic_web,
         "Open 24 Hours":Clinic_hours,
      })



data.to_excel('Scraped_data.xlsx', index=False)
```
